## 3. ***Data Wrangling***

### Data Wrangling Code

**Clean Column Names**

````python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_')     #Cleaning our variables
df.replace(['null', 'None', '', ' '], np.nan, inplace=True)                      #Replace inconsistent nulls with proper NaN values
````

**Convert Date Columns**

````python
date_cols = [col for col in df.columns if 'date' in col]
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')               #'errors="coerce"' ensures that invalid date formats are replaced with NaT
````

**Convert Numeric Columns**

````python
for col in df.select_dtypes(include=['object']).columns:

    try:                        # Try converting to numeric where possible
        df[col] = pd.to_numeric(df[col])
    except:
        continue
````

**Handle Duplicates**

````python
df.drop_duplicates(inplace=True)             # Check for duplicates
````

**Handle Missing Values**

Drop columns

````python
missing_percentage = df.isnull().mean()          #Drop columns with more than 50% missing values
drop_cols = missing_percentage[missing_percentage > 0.5].index
df.drop(columns=drop_cols, inplace=True)
````

Fill numeric missing values with median

````python
num_cols = df.select_dtypes(include=[np.number]).columns          #replace missing values (NaN) with median of that column
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
````

Fill categorical missing values with mode

````python
cat_cols = df.select_dtypes(include=['object']).columns               # Selecting columns in dataframe that have categorical (object/string) data types
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
````

**Standardize Categorical Columns**

````python
df['country'] = df['country'].str.strip().str.title()              #Clean country names & shipment modes
df['shipment_mode'] = df['shipment_mode'].str.strip().str.title()
df['vendor_inco_term'] = df['vendor_inco_term'].str.strip().str.upper()
````

Feature **Engineering**

````python
# Delivery delay
df['delivery_delay_days'] = (df['delivered_to_client_date'] - df['scheduled_delivery_date']).dt.days    #in days

# Delivery status
df['delivery_status'] = np.where(df['delivery_delay_days'] > 0, 'Delayed', 'On Time')
````

**Reset Index & Final Clean**

````python
df.reset_index(drop=True, inplace=True)
````

**Save Cleaned Data**

````python
df.to_csv("cleaned_fedex_logistics.csv", index=False)
print("Data wrangling complete! \nCleaned file saved as 'cleaned_fedex_logistics.csv'")
````

##### 1. Why did you pick the specific chart?

The pair plot, also known as a scatterplot matrix, is one of the most insightful tools in exploratory data analysis (EDA). It enables analysts to visualize pairwise relationships between multiple numeric variables in a single view.

In the context of logistics and supply chain analysis, our dataset includes several important numeric variables such as freight cost, insurance cost, item quantity, weight, and delivery delay days. These variables are all potentially interrelated.

While a single scatterplot can only show the relationship between two variables, a pair plot allows us to explore all possible variable combinations simultaneously. Each scatterplot in the grid represents the relationship between a pair of variables, and the diagonal typically shows the distribution (histogram or KDE) of each variable individually.

This chart was chosen for its clarity and comprehensiveness. It provides both individual variable distributions and cross-variable interactions. By analyzing this matrix, we can easily identify:

Linear or non-linear relationships among variables,

Clusters, outliers, or trends, and

Potential correlations that may influence cost, efficiency, or performance.
